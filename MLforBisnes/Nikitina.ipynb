````python
# Вопрос 2
# В формуле расчета сигмы в знаменателе корень из n => при увеличении n, стандартное отклонение будет уменьшаться, 
# поэтому с уменьшением сигмы интервал становится уже

# Проверка
import statsmodels.api as sm

k = 479

for n in [1000, 2000]:
    print(
        f'доверительный интервал, при n={n} ',
        sm.stats.proportion_confint(count=k, nobs=n, alpha=(1 - 0.95), method='normal'))
````
