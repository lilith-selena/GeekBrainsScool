импротируем библиотеки

```python
from pprint import pprint
from lxml import html
import requests
import pandas as pd
```

парсинг будет выполнятся сайта https://lenta.ru/

для того чтоб сайт не заблокировал пропишем актуальный User-Agent

спарсим информацию с сайта и переведем ее в text формат

```python
url = 'https://lenta.ru/'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 YaBrowser/23.1.2.928 Yowser/2.5 Safari/537.36',
}
response = requests.get(url=url, headers=headers)
response.text
```

представим информацию в формате DOM - объектная модель документа

```python
dom = html.fromstring(response.text)
dom
```

посмотрим сколько ссылок с новостями получилось

```python
link_list = dom.xpath('//section/div/div/div/a')
len(link_list)
```

посмотрим на них

```python
link_list
```

Получим ссылки на новости.
Ссылки которые ведут на ресурс https://lenta.ru/ сайт предоставил без начального адреса. Чтоб ими в дальнейшем востпользоваться необходимо прописывать в начале строки https://lenta.ru. Ссылки на внешнии ресурсы сразу кликабельны.

```python
link_ = dom.xpath('//section/div/div/div/a/@href')
link_
```

Заголовк на сайте спрятаны с помощью функции text(), принстскрины приложены в файле PrintScrin.doxc

```python
link_name= dom.xpath('//section/div/div/div/a/div/span/text()')
link_name
```

таким же образом получим время

```python
link_time= dom.xpath('//section/div/div/div/a/div/div/time/text()')
link_time
```

для удобства сопоставления информации переведем все списки в DataFrame и соединим по элементно. итоговый dataframe представлен ниже

```python
link = pd.DataFrame({'ссылка':link_})
name = pd.DataFrame({'тема':link_name})
time = pd.DataFrame({'дата публикации':link_time})

chek = pd.concat([link, name, time],axis=1)
chek
```
