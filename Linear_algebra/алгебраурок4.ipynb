__1.__ Решите систему уравнений методом Гаусса:

$$\begin{cases}

x_{1}+x_{2}-3x_{3}+x_{4}=4.
\end{cases}$$

$$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
-x_{2}-x_{3}+5x_{4}=-2, \\
-2x_{3}+3x_{4}=4.
\end{cases}$$

Получается неизвестных больше чем уравнений, решу в общем виде систему, для этого $x_{4}=c$ .Проходя обратный ход метода Гаусса, последовательно подставляя значения в уравнения, получу:

$$x_{3}=1,5с - 2\\
x_{2}=6,5с \\
x_{1}=-3с - 2$$

# Системы линейных уравнений. Часть 2

__1.__ Решите систему уравнений методом Крамера:

   а) $\begin{cases}
x_{1}-2x_{2}=1 \\
3x_{1}-4x_{2}=7
\end{cases}$

Найдем определитель матрицы коэффициентов:  

$A=\begin{pmatrix}1 & -2\\
3 & -4\end{pmatrix}\\
detA=\begin{vmatrix}1 & -2\\
3 & -4\end{vmatrix}=1\cdot -4-(-2)\cdot 3=2\neq 0\\
\Rightarrow\text{ Система определена}\\
detA_1=\begin{vmatrix}1 & -2\\
7 & -4\end{vmatrix}=-4+14=10\\
detA_2=\begin{vmatrix}1 & 1\\
3 & 7\end{vmatrix}=7-3=4\\
x_1=\frac{detA_1}{detA}=\frac{10}{2}=5\\
x_2=\frac{detA_2}{detA}=\frac{4}{2}=2$  


   б) $\begin{cases}
2x_{1}-x_{2}+5x_{3}=10 \\
x_{1}+x_{2}-3x_{3}=-2 \\
2x_{1}+4x_{2}+x_{3}=1
\end{cases}$



Найдем определитель матрицы коэффициентов:  

$$detA=\begin{vmatrix}

$$detA_{1}=\begin{vmatrix}

$$detA_{2}=\begin{vmatrix}
